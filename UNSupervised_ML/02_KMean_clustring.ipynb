## **KMean Clustring Algorithm**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
from sklearn.datasets import make_blobs
```

```python
x , y = make_blobs(n_samples=1000 , centers=3 , n_features=2)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[: , 0] , x[: , 1] , c=y)
```

```python
# standardization  -- feature scaling technique
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.33 ,random_state=42)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# apply Kmean clustring
from sklearn.cluster import KMeans
```

```python
# apply Elvow method to select k Value
wcss = []
for k in range(1 , 11):
    kmeans = KMeans(n_clusters=k , init="k-means++")
    kmeans.fit(x_train_scaled) # only train the data
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# plot elbow curve
plt.plot(range(1,11) ,wcss)
plt.xticks(range(1,11) )
plt.xlabel("Number of Clustres")
plt.ylabel("WCSS")
plt.show()
```

NOte:- we show the sharp edge on the the 3 so the cluster 3
k = 3

```python
kmeans =  KMeans(n_clusters=3 , init="k-means++")
kmeans.fit_predict(x_train_scaled) # this line train + label output do in single line
```

```python
y_pred = kmeans.predict(x_test_scaled)
y_pred
```

```python
# predict cluster for new value(s)
new_values = np.array([[0.5, -2.1], [3.2, 1.7]])
new_values_scaled = scaler.transform(new_values)
new_value_clusters = kmeans.predict(new_values_scaled)

print("Predicted cluster(s):", new_value_clusters)
```

```python
plt.scatter(x_test[: ,0] , x_test[ : , 1] , c=y_test)
```

secound way for validating the k value

```python
from sklearn.metrics import silhouette_score
silhouette_Coeff = [ ]
for k in range(2,11):
    kmeans = KMeans(n_clusters= k  , init= "k-means++")
    kmeans.fit(x_train_scaled)
    scour = silhouette_score(x_train_scaled , kmeans.labels_)
    silhouette_Coeff.append(scour)
```

```python
silhouette_Coeff
```

NOte: we chose the value where the point high the go down

```python
# plot elbow curve
plt.plot(range(2,11) ,silhouette_Coeff)
plt.xticks(range(2,11) )
plt.xlabel("Number of Clustres")
plt.ylabel("silhouette_Coeff")
plt.show()
```
