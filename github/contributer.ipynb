Ek dusre ki repository (Open Source ya Team Project) mein contribute karne ka process bilkul ek **"Naye Warehouse"** se saaman laane aur apna box wahan bhejne jaisa hai.

Isse hum Git ki bhasha mein **Forking and Pull Request (PR) Workflow** kehte hain. Chaliye isse step-by-step samajhte hain:

---

### Step 1: Forking (Apni Copy Banana)
Sabse pehle aap GitHub par us project ke page par jaate hain aur **"Fork"** button dabate hain.
* **Table-Box Example:** Ye aisa hai jaise aapne kisi aur ke Warehouse ki ek "Photo Copy" apne naam par kar li. Ab aap us copy mein jo chahe badlaav kar sakte hain, asli warehouse ko kuch nahi hoga.

### Step 2: Clone (Local Setup)
Ab aap us forked repository ko apne computer (Desk) par laate hain.
```bash
git clone https://github.com/YourUsername/ProjectName.git
```
* **Working:** Warehouse ki copy se saara saaman utha kar apne **Desk (Local Machine)** par rakhna.

### Step 3: Create a Branch (Naya Kaam Shuru Karna)
Hamesha naye kaam ke liye ek alag branch banayein:
```bash
git checkout -b my-new-feature
```
* **Working:** Ye aisa hai jaise aapne apne desk par ek **Alag Section** bana liya sirf "New Feature" ke liye, taaki main code kharab na ho.



### Step 4: Make Changes & Stage (Packing Table)
Ab aap files edit karte hain aur unhe packing table par rakhte hain:
```bash
git add .
```
* **Working:** Jo files aapne badli hain, unhe **Packing Table (Staging Area)** par rakh diya.

### Step 5: Commit (Box Seal Karna)
Apne badlaav ko ek box mein seal karein:
```bash
git commit -m "Added a cool new feature"
```
* **Working:** Packing table se saaman utha kar ek **Sealed Box (Commit)** bana diya aur us par ek note likh diya.

### Step 6: Push (Apne Warehouse bhejna)
Ab ye box apne online repository (Forked repo) par bhejein:
```bash
git push origin my-new-feature
```
* **Working:** Apne local desk se box ko apne **Personal Warehouse (GitHub Fork)** mein bhej dena.

### Step 7: Pull Request (Asli Warehouse ko Request bhejna)
Ab aap GitHub par jaakar **"New Pull Request"** par click karte hain.
* **Working:** Ye sabse important step hai. Aap asli project ke owner ko ek **Letter (Request)** bhej rahe hain: *"Maine aapke warehouse ki copy mein ye 1 box (commit) add kiya hai, kya aap ise apne Main Warehouse mein shamil karenge?"*



---

### 💡 Ek Pro Tip: Keeping Sync (Upstream)
Agar asli repository mein koi naya kaam hua hai aur aapko wo apne desk par chahiye, toh hum **Upstream** set karte hain:

1. **`git remote add upstream <Original-Repo-URL>`**: Asli warehouse ka pata (address) note karna.
2. **`git pull upstream main`**: Asli warehouse se naye boxes direct apne desk par mangwana.

**Summary Table:**

| Step | Command | Warehouse Example |
| :--- | :--- | :--- |
| **Fork** | (GitHub UI) | Dusre ke Warehouse ki Copy banana |
| **Clone** | `git clone` | Copy ko apne Desk par lana |
| **Branch** | `git checkout -b` | Desk par ek alag section banana |
| **Stage** | `git add` | Saaman Packing Table par rakhna |
| **Commit** | `git commit` | Box seal karke note likhna |
| **Push** | `git push` | Box apne Warehouse mein bhejna |
| **PR** | (GitHub UI) | Asli owner se box accept karne ki request karna |

Kya aap kisi specific open-source project par contribute karne ka plan kar rahe hain? Main aapko **`git remote add upstream`** ko setup karne ka exact tarika bata sakta hoon.
