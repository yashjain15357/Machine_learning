Pulling a Docker image is the process of downloading a pre-built image from a registry (like Docker Hub) to your local machine. Think of it like downloading an app from an app store so you can run it whenever you want.

### The Standard Pull Process

To pull an image, you use the `docker pull` command followed by the name of the image.

```bash
docker pull <image_name>:<tag>
```



---

### Step-by-Step Breakdown

1.  **Identify the Image:** Find the image you need on [Docker Hub](https://hub.docker.com). For example, if you need a web server, you might look for `nginx`.
2.  **Choose a Tag (Optional):** Tags represent versions. If you don't specify one, Docker defaults to `:latest`.
    * *Example:* `docker pull node:18-alpine` (pulls version 18 of Node.js on a lightweight Linux base).
3.  **Run the Command:** Enter the command in your terminal.
4.  **Layer Download:** Docker images are made of layers. You will see several bars on your screen as Docker downloads each layer. If you already have some of these layers from a different image, Docker will intelligently skip them to save time.
5.  **Verification:** Confirm the image is now on your system by listing your local images.

---

### Key Commands to Know

| Action | Command | Description |
| :--- | :--- | :--- |
| **Pull Image** | `docker pull nginx` | Downloads the latest Nginx image. |
| **List Images** | `docker images` | Shows all images currently stored on your PC. |
| **Inspect** | `docker inspect nginx` | Shows low-level information (size, OS, layers) of the image. |
| **Remove** | `docker rmi nginx` | Deletes the local image to free up space. |

---

### Pro Tip: Pulling vs. Running
You don't *always* have to run `docker pull` manually. If you run `docker run hello-world` and you don't have that image locally, Docker is smart enough to perform the **pull** automatically before it starts the container.
