Creating a Docker image is like writing a recipe. You define the ingredients (files and dependencies) and the cooking instructions (commands) in a file, then you "bake" it into an image.

Here is the step-by-step workflow:

### 1. Create a Project Directory
First, organize your files. Create a folder for your project and move into it.
```bash
mkdir my-docker-app && cd my-docker-app
```

### 2. Create your Application
For this example, imagine a simple Python script or a basic HTML file. Let’s say you have a file named `app.py`.

### 3. Create the Dockerfile
The **Dockerfile** is a text document containing all the commands a user could call on the command line to assemble an image. Create a file named `Dockerfile` (no file extension) and add your instructions:

```dockerfile
# Step 1: Choose a base image (the OS/environment)
FROM python:3.9-slim

# Step 2: Set the working directory inside the container
WORKDIR /app

# Step 3: Copy your local files into the container
COPY . /app

# Step 4: Run any necessary setup commands
RUN pip install --no-cache-dir -r requirements.txt

# Step 5: Define the command to run your app
CMD ["python", "app.py"]
```

### 4. Build the Image
Use the `docker build` command. The `-t` flag allows you to "tag" your image with a friendly name so you can find it later.

```bash
docker build -t my-custom-app:v1 .
```
> **Note:** The `.` at the end is crucial; it tells Docker to look for the Dockerfile in the current directory.



---

### 5. Verify the Image
Once the build finishes, check that your image is stored locally:
```bash
docker images
```

### 6. Run a Container from your Image
Now that the image exists, you can start a container based on it:
```bash
docker run my-custom-app:v1
```

---

### Summary Table: The Build Lifecycle

| Step | Command | What it does |
| :--- | :--- | :--- |
| **Write** | `nano Dockerfile` | Define the image blueprint. |
| **Build** | `docker build -t <name> .` | Execute the blueprint to create the image. |
| **Check** | `docker images` | Confirm the image exists in local storage. |
| **Run** | `docker run <name>` | Instantiate the image into a running container. |
