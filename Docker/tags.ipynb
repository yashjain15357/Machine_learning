In Docker, a **Tag** is essentially an alias or a version label for an image. Without tags, you would have to remember long, random IDs like `sha256:e9d105...` to run your containers.

Think of an **Image** as the "App" and the **Tag** as the "Version Number" (e.g., `v1.0`, `beta`, or `latest`).

---

### Use of Tags (Why use them?)

1.  **Version Control:** Allows you to keep multiple versions of the same app (e.g., `my-app:1.0` and `my-app:2.0`).
2.  **Environment Separation:** You can label images for different stages, like `my-app:development` or `my-app:production`.
3.  **Rollbacks:** If a new update breaks your site, you can quickly switch back to an older tag (e.g., `v1.5`).
4.  **Clarity:** The `:latest` tag is the default, but specific tags (like `:18-alpine` for Node.js) tell you exactly what’s inside the image (version 18, lightweight Linux).

---

### Steps to Use Tags

#### 1. Tagging During a Build
The most common way to tag an image is while you are creating it.
```bash
# Syntax: docker build -t <name>:<tag> .
docker build -t my-website:v1.0 .
```

#### 2. Tagging an Existing Image
If you have an image already and want to give it a new name or version (often used before pushing to Docker Hub):
```bash
# Syntax: docker tag <source-image>:<tag> <target-name>:<new-tag>
docker tag my-website:v1.0 yourusername/my-website:latest
```

#### 3. Running a Specific Tag
If you don't specify a tag, Docker assumes you want `:latest`. To run a specific version:
```bash
docker run my-website:v1.0
```

#### 4. Pushing a Tag to a Registry
When you push to Docker Hub, the tag tells the registry where to store that specific version.
```bash
docker push yourusername/my-website:v1.0
```

---

### Pro-Tip: The "latest" Trap
The `:latest` tag is just a name; it **does not** automatically mean it is the most recent code you wrote. It simply points to the most recent build that was labeled "latest." 

> **Best Practice:** Always use specific version numbers (like `v1.2.3`) for production so you know exactly what code is running.

### Summary of Commands

| Goal | Command |
| :--- | :--- |
| **List all tags** | `docker images` |
| **Create new tag** | `docker tag old-image:v1 new-image:v2` |
| **Pull specific tag** | `docker pull python:3.9-slim` |
| **Remove a tag** | `docker rmi my-app:v1` |
