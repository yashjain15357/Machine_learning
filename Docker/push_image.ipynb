To push an image to Docker Hub, you first need a **Docker Hub account**. Pushing is essentially "uploading" your custom recipe so others (or your servers) can download and use it.

Here are the 4 essential steps to get your image from your local machine to the cloud.

---

### 1. Log In to Docker Hub
Before you can upload anything, you must authenticate your terminal with your Docker ID.

```bash
docker login
```
* **Username:** Enter your Docker Hub ID.
* **Password:** Enter your password (or a Personal Access Token if you have 2FA enabled).

### 2. Tag Your Image Correctly
Docker Hub requires a specific naming convention to know where to store the image. It must follow this format: `your-username/repository-name:tag`.

If you already built an image named `my-app`, you need to re-tag it:
```bash
# Syntax: docker tag <existing-image> <your-username>/<repo-name>:<version>
docker tag my-app:v1 yourusername/my-app:v1
```

### 3. Push the Image
Now that the image is properly named, you can upload it to the registry.

```bash
docker push yourusername/my-app:v1
```
> **What happens here:** Docker will upload the individual "layers" of your image. If Docker Hub already has some of those layers (like the base Linux OS), it will skip them to save time and bandwidth.

### 4. Verify on Docker Hub
Once the push is complete, you can:
1.  Go to [hub.docker.com](https://hub.docker.com).
2.  Log in and look under **Repositories**.
3.  You should see `my-app` listed there with the `v1` tag.

---

### Summary Checklist

| Step | Command | Why? |
| :--- | :--- | :--- |
| **1. Auth** | `docker login` | Connects your CLI to your cloud account. |
| **2. Rename** | `docker tag ...` | Prepares the image with your username prefix. |
| **3. Upload** | `docker push ...` | Sends the image layers to the registry. |
| **4. Check** | `docker images` | Confirm the new tag exists locally first. |

---

### Important Note on Privacy
By default, Docker Hub repositories are **Public**, meaning anyone in the world can pull your image. If your image contains sensitive code or API keys, make sure to set the repository to **Private** in the Docker Hub settings first!
