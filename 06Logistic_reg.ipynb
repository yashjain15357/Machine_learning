Logistic regression


```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
# create a classification dataset by the sklearn

from sklearn.datasets import make_classification
```

```python
# create a dataset
x,y=make_classification(n_samples=1000 , n_features=10 , n_classes=2 , random_state=42)
```

```python
pd.DataFrame(x)
```

```python
#  train test split
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x , y , train_size=0.75, random_state=42)
```

```python
# Model training
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(x_train , y_train)
y_pred = logistic.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

scoure = accuracy_score(y_test , y_pred)
print(scoure)
cm = confusion_matrix(y_test , y_pred)
print(cm)
```

```python
# hyper parameter tuning and cross validation
# use for check the best parameter combination 
# great search 
# random search cross validation

model = LogisticRegression()
penalty = ['l1' ,'l2','elasticnet']
c_value = [100, 10, 1.0, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
param = dict(penalty = penalty , C=c_value , solver = solver )

```

```python
from sklearn.model_selection import StratifiedKFold
cv= StratifiedKFold(n_splits=5,shuffle=False)
```

```python
##GrideSearch CV

from sklearn.model_selection import GridSearchCV
grid =GridSearchCV(estimator=model , cv=5 , param_grid=param , scoring= 'accuracy',n_jobs= -1)
grid

```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

### Random search cv

```python
from sklearn.model_selection import RandomizedSearchCV
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=param,cv=5,scoring='accuracy')
randomcv.fit(x_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred = randomcv.predict(x_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Logistic Regression For Multiclass Classification Problem

```python
# create a classification dataset by the sklearn

from sklearn.datasets import make_classification
# create 3 classes data set

x,y = make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
from sklearn.model_selection import train_test_split 
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size=0.70, random_state= 42 )

```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(multi_class='ovr')
```

```python
logistic.fit(x_train , y_train  )
y_pred= logistic.predict(x_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
