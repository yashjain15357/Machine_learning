## ML INTRODUCTION

Machine Learning (ML) is a part of Artificial Intelligence that enables a computer to learn from data and improve its performance without being explicitly programmed.

There are **mainly three types of Machine Learning**:

1. **Supervised Learning**
   → Data is labeled (input + output given)

2. **Unsupervised Learning**
   → Data is unlabeled (only input given)

3. **Reinforcement Learning**
   → Model learns by reward and punishment


### ✅ Supervised Learning (Hinglish – for your notes)

**Supervised Learning** ek aisa Machine Learning type hota hai jisme
👉 hume pehle se **labeled data** diya hota hai
(matlab input ke saath correct output bhi hota hai).

Model ko pehle training data par sikhaya jata hai,
phir wo naye data ka **output predict** karta hai.

### 📌 Example
Maan lo tumhare paas data hai:
| Area (sqft) | Price (₹) |
| ----------- | --------- |
| 800         | 30 lakh   |
| 1000        | 40 lakh   |
| 1200        | 50 lakh   |

Yaha
👉 **Area = input**
👉 **Price = output (label)**

Model in data se sikhega aur phir bolega:

👉 1100 sqft ka ghar ho to approx kitni price hogi.



### ✅ Unsupervised Learning

**Unsupervised Learning** ek aisa Machine Learning type hota hai jisme
👉 hume **sirf input data milta hai**,
👉 **output / label nahi diya hota**.

Model khud hi data ke andar
👉 **patterns, groups ya structure** find karta hai.

**Simple words mein:**
👉 Only input, no correct answer given = Unsupervised Learning.

---
### 📌 Example
Maan lo tumhare paas customer ka data hai:
| Age | Income |
| --- | ------ |
| 22  | 20k    |
| 25  | 22k    |
| 45  | 60k    |
| 48  | 65k    |

Yaha koi bataya nahi hai ki kaun sa customer kis group ka hai.

Model khud se customers ko
👉 similar nature ke groups me baant deta hai
(jese young-low income, old-high income).

Isko kehte hain **Unsupervised Learning**.



### ✅ Reinforcement Learning (Hinglish – for your notes)

**Reinforcement Learning** ek aisa Machine Learning type hai jisme
👉 model (agent) environment ke saath interact karta hai
aur
👉 **reward ya punishment** ke basis par seekhta hai.

Model ko pehle se correct answer nahi diya hota,
wo try–try karke sikhata hai.

---

### 🔹 Simple definition

👉 Action lo → reward milta hai → model improve karta hai
= Reinforcement Learning.

---

### 🔹 Example

Maan lo ek game hai:

Agent = game player
Action = left / right move
Reward = score increase

Agar sahi move karega → reward milega
galat move karega → kam reward ya punishment

Dheere-dheere agent best strategy seekh leta hai.


Tum ML (especially **linear model / SVM / regression**) me jo decision boundary hoti hai usko hi

👉 **2D me line**,
👉 **3D me plane**,
👉 aur
👉 higher dimension me **hyperplane** kehte hain.

---

## ✅ 1. Line (2-D case)

Jab sirf **2 features** hote hain → boundary ek **line** hoti hai.

### ✔ Equation (2D)
$$
[
w_1 x + w_2 y + b = 0
]
$$
👉 x , y = features
👉 w₁ , w₂ = weights
👉 b = bias

Is equation se 2D graph me ek **straight line** banti hai.

---

## ✅ 2. Plane (3-D case)

Jab **3 features** hote hain → boundary ek **plane** hoti hai.

### ✔ Equation (3D)

[
w_1 x + w_2 y + w_3 z + b = 0
]

👉 x , y , z = features
👉 ye 3D space me ek **flat surface (plane)** hota hai.

---

## ✅ 3. Hyperplane (n-D case)

Jab features 3 se zyada ho jaate hain
(jaise x₁, x₂, x₃, x₄, …)

to boundary ko **hyperplane** kehte hain.

### ✔ General equation (n-dimension)

[
w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b = 0
]

Short form:
$$
[
\mathbf{w}^T \mathbf{x} + b = 0
]
$$
---



## ✅ Instance-based Learning

👉 Instance-based learning mein **model training ke time koi proper model build nahi karta**.
👉 Ye **poora training data ya instances store** karke rakhta hai.
👉 Jab new input aata hai, to **similar old data se compare karke answer deta hai**.

**Simple line:**
👉 Learn by remembering data.
### ✔ Example

**K-Nearest Neighbors (KNN)**

New point aaya → nearest points dekhe → majority se output.

### ✔ One line note

Instance-based learning mein prediction stored training instances ke basis par hota hai.

---

## ✅ Model-based Learning

👉 Model-based learning mein
**training data se ek mathematical model seekha jata hai.**
👉 Phir future prediction us learned model se hota hai.

**Simple line:**
👉 Learn by building a model.

### ✔ Example

Linear Regression, Logistic Regression, Decision Tree, Random Forest


### ✅ Best Fit Line (Hinglish – exam & notes ready)

**Best fit line** wo straight line hoti hai
jo given data points ke beech se aisi pass ho
ki **prediction error minimum** ho jaaye.

Simple words me:
👉 jo line data ko sabse achha represent kare
use **best fit line** kehte hain.

---

### 🔹 Mathematical form
$$
[
\hat{y} = wx + b
]
$$
---

### 🔹 Best fit ka matlab kya hota hai?

Best fit ka matlab ye nahi ki line sabhi points ko touch kare,
balki ye hota hai ki:

👉 actual value (y) aur predicted value (ŷ) ka difference
overall sabse kam ho.

---

### 🔹 Kaunsa error minimize hota hai?

Normally best fit line find karte waqt:
$$
[
\sum (y - \hat{y})^2
]
$$
yaani **Mean Squared Error (MSE)** minimize kiya jata hai.

---

### 🔹 Easy example

Maan lo tumne area vs price ka data liya.

Model ek line banata hai jisse:

* zyada tar points us line ke paas ho
* overall distance kam ho

👉 wahi line = best fit line
