## Polynomial Regression Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= 6 * np.random.rand(100 , 1) -3
y= 0.5 * x ** 2 + 1.5 * x + np.random.randn(100 , 1)
plt.scatter(x=x ,y=y)
# print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train1 , x_test1 , y_train, y_test = train_test_split(x , y ,test_size=0.25,random_state=42)
print(x_train1.shape)
```

```python
print(x_test1.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaller = StandardScaler()
x_train = scaller.fit_transform(x_train1)
x_test = scaller.transform(x_test1)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train , y_train)
```

```python
from sklearn.metrics import r2_score
ypred = regression.predict(x_test)
score = r2_score(y_test ,ypred)
print(score)
```
