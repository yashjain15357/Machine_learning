# Decision Tree Machine algorithm

#### ✅ Decision Tree kya hota hai? (Hinglish)

**Decision Tree** ek *non-parametric supervised learning algorithm* hota hai jo **classification** aur **regression** dono ke liye use hota hai.

Iska structure bilkul **tree (flowchart)** jaisa hota hai — jisme:

* upar **root node** hota hai
* beech me **decision / internal nodes** hote hain
* aur last me **leaf nodes** hote hain

Model feature-based splits karta hai aur end me **leaf node par final decision** milta hai.

---

#### 🌳 Decision Tree – Basic Idea

Decision Tree ek aisa tool hai jo alag-alag fields me use hota hai.
Ye data ko step-by-step split karta hai (features ke basis par) aur finally prediction deta hai.

---

### ✅ Types of Decision Tree

**1. ID3**
ID3 algorithm **entropy** ka use karta hai.
Ye check karta hai ki data kitna mixed hai aur fir wo feature choose karta hai jo data ko sabse achhe se separate kare.

**2. C4.5**
C4.5, ID3 ka improved version hai.
Ye:

* missing data handle kar sakta hai
* continuous attributes par bhi kaam karta hai

**3. CART**
CART algorithm **Gini impurity** use karta hai.
Ye:

* classification (category predict)
* aur regression (continuous value predict)
  dono ke liye use hota hai.

---


### What is Entropy?

**Entropy** is a measure of the impurity or disorder in a set of examples. In the context of decision trees, it quantifies how mixed a set of labels (in our case, "Yes" or "No" for "PlayTennis") is.

* **High entropy:** The set is highly impure, meaning there's an even mix of "Yes" and "No" outcomes. It's difficult to predict the outcome.
* **Low entropy:** The set is relatively pure, meaning most outcomes are either "Yes" or "No." It's easier to predict the outcome.
* **Zero entropy:** The set is completely pure, meaning all outcomes are the same (either all "Yes" or all "No"). The outcome is perfectly predictable.

---

### The Formula for Entropy

The mathematical representation of Entropy is:

$$Entropy(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

**Where:**
* $S$: The set of examples.
* $c$: The number of classes (in our case, "Yes" and "No," so $c=2$).
* $p_i$: The proportion of examples belonging to class $i$.

Gini Impurity, Decision Trees mein ek bahut hi popular metric hai jo humein ye batata hai ki hamara data kitna "ganda" ya "mixed" hai. Simple words mein, ye measure karta hai ki agar hum kisi random element ko dataset se uthayein, toh uske galat classify hone ke kitne chances hain.

Decision Trees (jaise **CART** algorithm) iska use karte hain taaki wo best "split" choose kar sakein.

---

### Gini Impurity Kya Hai?

Gini Impurity ki value **0 se 0.5** ke beech mein hoti hai (Binary classification ke liye):

* **Gini = 0:** Iska matlab hai ki set **"Pure"** hai. Saare examples ek hi class ke hain (Ya toh saare "Yes" hain, ya saare "No").
* **Gini = 0.5:** Iska matlab hai ki set **"Maximally Impure"** hai. Data bilkul 50-50 split hai, aur prediction karna sabse mushkil hai.

---

### Gini Impurity ka Formula

Gini Impurity ka formula kafi simple hota hai:

$$Gini(S) = 1 - \sum_{i=1}^{c} (p_i)^2$$

**Yahan par:**

* $S$ = Dataset.
* $c$ = Number of classes (e.g., "Yes" aur "No").
* $p_i$ = Us class ki probability ya proportion.

---

### Gini vs Entropy: Farq Kya Hai?

Aapne pehle Entropy sikha tha, toh ab confusion ho sakta hai ki dono mein se kaunsa better hai?

| Feature | Gini Impurity | Entropy |
| --- | --- | --- |
| **Complexity** | Simple hai, calculation fast hoti hai. | Thoda complex hai kyunki ismein `log` use hota hai. |
| **Speed** | Decision Trees jaldi build hote hain. | Thoda slow hota hai. |
| **Usage** | Default in Scikit-Learn (CART algorithm). | Used in ID3 / C4.5 algorithms. |
| **Range** | 0 to 0.5 (for 2 classes). | 0 to 1. |

---

### Conclusion: Gini Kyun Use Karein?

Machine learning models (jaise Random Forest) mein Gini Impurity isliye use hota hai kyunki ismein **computationally expensive log calculations nahi hote**. Result dono ke almost same aate hain, par Gini thoda fast kaam karta hai.

Kya aap chahte hain ki main **Outlook** attribute ka Gini split calculate karke dikhaun taaki aap compare kar sakein?

Machine Learning mein Decision Tree build karte waqt, **Gini Impurity** ek benchmark ki tarah kaam karta hai. Ye tree ko decide karne mein madad karta hai ki kaunsa sawaal (feature) sabse pehle poochna chahiye.

Aaiye isse detail mein samajhte hain:

---

### 1. Decision Tree mein Gini ka Kaam

Jab Decision Tree banta hai, toh uska main goal hota hai **"Pure Nodes"** banana.

* Agar kisi node mein saare "Yes" hain, toh impurity **0** hai.
* Agar data mixed hai, toh impurity high hai.

**CART (Classification and Regression Trees)** algorithm Gini Impurity ka use karta hai ye dekhne ke liye ki split karne ke baad data kitna "saaf" (pure) hua.

---

### 2. Best Split Kaise Chunta Hai? (Gini Gain)

Decision Tree har feature (Outlook, Temp, Humidity) ko check karta hai aur niche diye gaye steps follow karta hai:

1. **Har Branch ki Gini Nikalna:** Maan lo humne "Outlook" par split kiya. Toh hum Sunny, Overcast, aur Rain ki alag-alag Gini calculate karenge.
2. **Weighted Gini:** Phir hum poore feature ki ek average Gini nikalte hain (jisme zyada rows wale branch ko zyada weightage milti hai).
3. **Gini Impurity Kam Karna:** Tree us feature ko root node banata hai jiske wajah se **Gini Impurity sabse zyada girti hai**.




### ✅ Decision Tree Terminologies (Hinglish)

**Root Node**
Tree ka starting node hota hai, jahan se poora dataset split hona start hota hai.

**Decision Node**
Jo nodes root se split hone ke baad bante hain, unhe decision nodes kehte hain.
Ye beech ke decision represent karte hain.

**Leaf Node (Terminal Node)**
Jahan aage splitting nahi hoti.
Yahi par final output / class milti hai.

**Sub-Tree**
Decision tree ka koi chhota part ya portion, jise alag tree ki tarah dekha ja sakta hai.

**Branch / Sub-Tree**
Tree ka ek particular path (decisions ka sequence) branch ya sub-tree kehlata hai.

**Parent Node & Child Node**

* Jo node split hota hai → **Parent node**
* Usse nikle hue nodes → **Child nodes**

Parent node decision represent karta hai, aur child nodes us decision ke possible results hote hain.


In Machine Learning, **Information Gain** is the mathematical heartbeat of the ID3 algorithm. It is essentially the "Profit" you make in terms of clarity when you ask a specific question.

Think of it this way: You start with a messy, confused pile of data (**Entropy**). When you split that data using a feature (like "Outlook"), the pile becomes more organized. Information Gain is the measure of how much that messiness decreased.

---

### 1. The Core Concept (Hinglish Style)

Information Gain humein batata hai ki kisi feature ne hamari **"Bechaini" (Uncertainty)** kitni kam ki.

* **High Information Gain:** Iska matlab hai ki us feature ne data ko bahut acche se divide kiya (Pure nodes banaye).
* **Low Information Gain:** Iska matlab hai ki split karne ke baad bhi data kafi mixed hai, aur humein koi khaas faida nahi hua.

---

### 2. The Formula

Information Gain nikalne ke liye hum **Total Entropy** mein se **Weighted Average Entropy (Split ke baad wali)** ko minus karte hain.

$$Gain(S, A) = Entropy(S) - \sum_{v \in Values(A)} \frac{|S_v|}{|S|} Entropy(S_v)$$

**Breakdown:**

* $Entropy(S)$: Split se pehle ki total impurity.
* $\sum ...$: Split karne ke baad bachi hui impurity (weighted average).
* $Gain$: In dono ka difference. Jitna bada difference, utna accha feature!

---

### 3. Step-by-Step Working in a Decision Tree

Decision Tree banate waqt, algorithm ye steps follow karta hai:

1. **Calculate Parent Entropy:** Sabse pehle root level par pure dataset ki Entropy nikalte hain.
2. **Test Every Feature:** Algorithm har feature (Outlook, Temperature, Humidity) ke liye "Information Gain" calculate karta hai.
3. **Choose the Winner:** Jis feature ka **Information Gain sabse zyada** hota hai, usko Decision Node bana diya jata hai.
4. **Repeat:** Ye process tab tak chalta hai jab tak har branch "Pure" (Entropy = 0) na ho jaye.

---

### 4. Gini vs. Information Gain

Aksar log confuse hote hain ki Information Gain use karein ya Gini Impurity.

| Feature | Information Gain | Gini Impurity |
| --- | --- | --- |
| **Metric used** | **Entropy** (Uncertainty). | **Gini** (Misclassification probability). |
| **Algorithm** | Used in **ID3**, C4.5. | Used in **CART** (Scikit-Learn). |
| **Math** | Uses `log` calculations (Slower). | Uses squares (Faster). |
| **Outcome** | Slightly more precise in some cases. | Preferred for large datasets due to speed. |

Step 1: Calculate the Entropy of the Target Variable ("PlayTennis")

Apply the entropy formula:
    $$ Entropy(PlayTennis) = - (\frac{9}{14} \log2(\frac{9}{14}) + \frac{5}{14} \log2(\frac{5}{14})) $$

       $$ \log2(\frac{9}{14}) \approx \log2(0.643) \approx -0.638 $$
       $$ \log2(\frac{5}{14}) \approx \log2(0.357) \approx -1.484 $$

    $$ Entropy(PlayTennis) = - (0.643 \times -0.638 + 0.357 \times -1.484) $$
    $$ Entropy(PlayTennis) = - (-0.410 + -0.529) $$
    $$ Entropy(PlayTennis) = - (-0.939) $$
    $$ Entropy(PlayTennis) \approx 0.939 \text{ bits} $$

Step 3: Calculate Information Gain for Each Attribute

Information Gain measures how much the entropy of the target variable is reduced by splitting the dataset on a particular attribute. We want to choose the attribute with the highest information gain as the root node of our decision tree.

The formula for Information Gain is:

$$ Gain(S, A) = Entropy(S) - \sum{v \in Values(A)} \frac{|Sv|}{|S|} Entropy(Sv) $$

Where:

   $$ S $$ is the original dataset.
   $$ A $$ is the attribute we are considering.
   $$ Values(A) $$ are the possible values of attribute $$ A $$.
   $$ Sv $$ is the subset of $$ S $$ for which attribute $$ A $$ has value $$ v $$.
   $$ |Sv| $$ is the number of examples in $$ Sv $$.
   $$ |S| $$ is the total number of examples in $$ S $$.
   $$ Entropy(Sv) $$ is the entropy of the subset $$ Sv $$.

We already calculated $$ Entropy(S) = Entropy(PlayTennis) \approx 0.939 $$ and the weighted average entropies for each attribute. So, we can directly calculate the Information Gain.

Information Gain (Outlook):
    $$ Gain(PlayTennis, Outlook) = Entropy(PlayTennis) - Entropy(Outlook, PlayTennis) $$
    $$ Gain(PlayTennis, Outlook) = 0.939 - 0.674 $$
    $$ Gain(PlayTennis, Outlook) \approx 0.265 $$

Information Gain (Temperature):
    $$ Gain(PlayTennis, Temperature) = Entropy(PlayTennis) - Entropy(Temperature, PlayTennis) $$
    $$ Gain(PlayTennis, Temperature) = 0.939 - 0.912 $$
    $$ Gain(PlayTennis, Temperature) \approx 0.027 $$

Information Gain (Humidity):
    $$ Gain(PlayTennis, Humidity) = Entropy(PlayTennis) - Entropy(Humidity, PlayTennis) $$
    $$ Gain(PlayTennis, Humidity) = 0.939 - 0.783 $$
    $$ Gain(PlayTennis, Humidity) \approx 0.156 $$

Information Gain (Wind):

    $$ Gain(PlayTennis, Wind) = 0.939 - 0.892 $$
    $$ Gain(PlayTennis, Wind) \approx 0.047 $$
