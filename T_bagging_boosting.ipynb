
## 1. Bagging (Bootstrap Aggregating)

Bagging ka main goal hota hai **Variance** (overfitting) ko kam karna. Iska sabse bada example **Random Forest** hai.

* **Kaise kaam karta hai?** Aapke paas ek bada data hai. Bagging mein hum us data ke alag-alag chhote "bags" (subsets) banate hain.
* **Parallel Processing:** Har bag par ek alag model (jaise Decision Tree) train hota hai. Ye saare models ek saath (parallel) kaam karte hain.
Bagging mein hum sirf model ko train nahi karte, balki "Bootstrap" ka use karte hain. Iska matlab hai ki har model ko data ka ek random subset milta hai (with replacement), taaki har model thoda alag seekhe. Agar sabko exact same data de denge, toh sabka output bhi same aayega aur ensemble ka koi fayda nahi hoga.
* **Voting:** End mein, sabka result liya jata hai. Agar classification hai toh "Majority Voting" hoti hai, aur regression hai toh "Average" nikala jata hai.

> **Example:** Maan lo 5 dosto ko ek movie ka trailer dikhaya gaya. Sabne apna-apna opinion diya. End mein jis movie ko sabse zyada "Hit" votes mile, wahi final result hai.

## 2. Boosting

Boosting ka main goal hota hai **Bias** (underfitting) ko kam karna aur model ko "strong" banana. Iske examples hain **XGBoost** aur **AdaBoost**.

* **Kaise kaam karta hai?** Ye parallel nahi, balki **Sequential** (ek ke baad ek) kaam karta hai.
* **Mistake Correction:** Pehla model train hota hai, jo galtiyan wo karta hai, doosra model unhi galtiyon ko sudhaarne par focus karta hai. Phir teesra model doosre ki galti sudhaarta hai.
* **Strong Learner:** Is tarah weak learners milkar ek bahut "Strong Learner" banate hain.

### Boosting Ka Step-by-Step Process:

1. **Model 1:** Poore data par train hota hai. Maan lo isne 100 mein se 80 sahi kiye aur 20 galat.
2. **Focus on Errors:** Agle step mein, hum un 20 galat points ko "highlight" kar dete hain (unka weight badha dete hain).
3. **Model 2:** Ab ye model un 20 galat points ko sahi karne ki koshish karta hai.
4. **Combination:** Isi tarah naye models bante jaate hain (Sequential). End mein, saare models ke results ko combine karke ek final strong prediction nikalte hain.

**"Increasing Weight" ka real matlab kya hai?:-**

ML model ke liye weight badhane ka matlab hai usse ye kehna: "Bhai, agar tune in 20 points ko galat predict kiya, toh tujhe bahut bada 'Penalty' (Loss) milega!"

> **Example:** Ye ek student jaisa hai jo test deta hai. Pehli baar mein jo sawal galat hue, wo agli baar unhi par zyada dhyan deta hai taaki score behtar ho sake.

## Key Differences: Ek Nazar Mein

| Feature | Bagging | Boosting |
| --- | --- | --- |
| **Workflow** | Parallel (Saath-saath) | Sequential (Ek ke baad ek) |
| **Focus** | Overfitting kam karna | Accuracy badhana/Galti sudharna |
| **Data Selection** | Random subsets | Galti waale data ko zyada weight milta hai |
| **Example** | Random Forest | XGBoost, CatBoost |
